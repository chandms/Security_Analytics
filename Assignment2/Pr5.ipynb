```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
mul = np.matmul
```

```python
opt_train = pd.read_csv('optdigits.tra',header=None)
```

```python
opt_train.head()
```

```python
X= pd.concat([pd.Series(1, index=opt_train.index, name='00'), opt_train.iloc[:,:64]], axis=1)
```

```python
y = opt_train.iloc[:,64]
```

```python
y1 = np.zeros([opt_train.shape[0], len(y.unique())])
y1 = pd.DataFrame(y1)
```

```python
for i in range(0, len(y.unique())):
    for j in range(0, len(y1)):
        if y[j] == y.unique()[i]:
            y1.iloc[j, i] = 1
        else: 
            y1.iloc[j, i] = 0
#y1.head()
```

```python
def sigmoid(X):
    return 1/(1+np.exp(-(X)))
```

```python
def hypothesis(X, theta):
    z = np.dot(X, theta)
    return sigmoid(z)
```

```python
def computeCost(theta, X, y, lamda=None):
    m = X.shape[0]
    if lamda:
        return (1/m) * mul(X.T, (hypothesis(X, theta) - y)) + (lamda/m) * theta 
    return (1/m) * mul(X.T, (hypothesis(X, theta) - y)) 
```

```python
def update_theta(theta, X, y, alpha, lamda=None):
    return theta - alpha * computeCost(theta, X, y, lamda)
```

```python
def gradient_descent(X, y, theta, alpha, epochs, lamda=None):
    m = len(X)
    for i in range(0, epochs):
        for j in range(0, 10):
            theta = pd.DataFrame(theta)
            h = hypothesis(X,theta.iloc[:,j])
            for k in range(0, theta.shape[0]):
                theta.iloc[k, j] = update_theta(theta.iloc[k, j],X.iloc[:, k],y.iloc[:, j],alpha,lamda)
            theta = pd.DataFrame(theta)
    return theta
```

```python
theta = np.zeros([X.shape[1], y1.shape[1]])
theta = gradient_descent(X, y1, theta, 0.01, 1500,lamda=0.5)
```

```python
output = []
for i in range(0, 10):
    theta1 = pd.DataFrame(theta)
    h = hypothesis(X,theta1.iloc[:,i])
    output.append(h)
output=pd.DataFrame(output)
```

```python
accuracy = 0
for col in range(0, 10):
    for row in range(len(y1)):
        if y1.iloc[row, col] == 1 and output.iloc[col, row] >= 0.5:
            accuracy += 1
accuracy = accuracy/len(X)
```
